```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def converter_para_cinza_manual(imagem_path, imagem):
    # Abrir a imagem
    imagem = Image.open(imagem_path + imagem)

    # Obter as dimensões da imagem
    largura, altura = imagem.size

    # Criar uma nova imagem em modo "RGB" para armazenar os valores de cinza
    exemplo_cinza = Image.new("RGB", (largura, altura))

    exemplo_preto_branco = Image.new("RGB", (largura, altura))

    # Processar cada pixel da imagem original
    for y in range(altura):
        for x in range(largura):
            # Obter o valor RGB do pixel
            r, g, b = imagem.getpixel((x, y))

            # Calcular o valor de cinza usando a fórmula ponderada
            cinza = int(0.299 * r + 0.587 * g + 0.114 * b)

            # Utiliza o limiar para definir a tonalidade de preto
            preto_branco = 255 if cinza > 127 else 0

            # Criar o pixel em tons de cinza (RGB com os três valores iguais)
            exemplo_cinza.putpixel((x, y), (cinza, cinza, cinza))
            exemplo_preto_branco.putpixel((x, y), (preto_branco, preto_branco, preto_branco))

    nome_imagem_cinza = "exemplo_cinza.jpeg"
    nome_imagem_preto_branco = "exemplo_preto_branco.jpeg"
    # Salvar a imagem em tons de cinza como JPEG
    exemplo_cinza.save(imagem_path + nome_imagem_cinza)
    exemplo_preto_branco.save(imagem_path + nome_imagem_preto_branco)

    # Exibir a imagem em tons de cinza
    exemplo_cinza.show()
    exemplo_preto_branco.show()
    print(f"Imagem convertida para tons de cinza e salva em {imagem_path}")
```

```python
def apresenta_imagem(original, cinza, preto):
    # Exibir as imagens lado a lado
    plt.figure(figsize=(15, 5))

    # Exibir imagem original
    plt.subplot(1, 3, 1)
    plt.imshow(original)
    plt.title("Imagem Original")
    plt.axis('off')

    # Exibir imagem em tons de cinza
    plt.subplot(1, 3, 2)
    plt.imshow(cinza)
    plt.title("Imagem em Tons de Cinza")
    plt.axis('off')

    # Exibir imagem em tons de preto
    plt.subplot(1, 3, 3)
    plt.imshow(preto)
    plt.title("Imagem em Tons de Preto")
    plt.axis('off')

    # Mostrar as imagens
    plt.show()
```
